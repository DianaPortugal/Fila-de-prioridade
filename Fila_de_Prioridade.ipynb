**Problema**

Uma fila de prioridade é um tipo abstrato de dados que mantém um conjunto S de elementos, cada qual com uma prioridade associada e que suporta as seguintes operações:

● Maximum(S): retorna o elemento de S que possui a maior prioridade.

● Extract-max(S): remove e devolve o elemento de S que possui a maior prioridade.

● Insert(S, x, prior): insere o elemento x que tem uma prioridade prior no conjunto S.


A estrutura de dados heap, por sua vez, é uma estrutura de dados que tem a seguinte propriedade: o conteúdo de um nó é maior ou igual que o conteúdo dos nós na subárvore enraizada nele. Um heap pode ser visualizado como uma árvore binária e representado por um vetor. A árvore é completamente preenchida em todos os níveis, exceto ossivelmente o último nível em que as folhas estão o mais à esquerda possível.

Neste trabalho você deve implementar uma fila de prioridade usando um max-heap.

```python
class ELEMENTO:
  def __init__(self, id,prior):
    self.id=id
    self.prior=prior

  def putPrior(self, Num_float):
    candidate = Num_float
    if (candidate >=0 and candidate <=900000) or candidate ==-1:
      self.prior = Num_float
    else:
      print("Valid for prior must follow the rule: 0.0 #0<=prior<=900000.0")
```

```python
class FILAPRIORIDADE():
  maxElementos=4000 #integer, that represents number of elements in A
  m =0 #actual number of elements on heap
  A = [None]*maxElementos #vector of elements

  def get_right_child(self, A ,index): #auxiliar do max_heapify e devolve o filho da direita para uma dada posição do vetor de prioridades
    if((((2*index)+1) < len(A)) and (index >= 1)):
      return (2*index)+1
    return -1

  def get_left_child(self, A, index): #auxiliar do max_heapify e devolve o filho da esquerda para uma dada posição do vetor de prioridades
    if(((2*index) < len(A)) and (index >= 1)):
        return 2*index
    return -1

  def max_heapify(self, A, index): # Auxiliar de decrease_key, extract_max, build_max_Heap - garante a propriedade do pai ter maior prioridade que os filhos a partir de um elemento
    left_child_index = self.get_left_child(A, index)
    right_child_index = self.get_right_child(A, index)

    largest = index
    if ((left_child_index <= self.m) and (left_child_index>0)):
      if (float(A[left_child_index].prior.replace(',','.')) > float(A[largest].prior.replace(',','.'))):
        largest = left_child_index
    if ((right_child_index <= self.m and (right_child_index>0))):
      if (float(A[right_child_index].prior.replace(',','.')) > float(A[largest].prior.replace(',','.'))):
        largest = right_child_index
    if (largest != index):
      A[index], A[largest] = A[largest], A[index]
      self.max_heapify(A, largest)

  def maximum(self): # Método implementa a opção 1 do menu - impressão do topo da fila de prioridades
    if self.m==0:
      return "-1 -1,0"
    else:
      return str(self.A[1].id) +" "+ str(float(self.A[1].prior.replace(',','.'))).replace('.',',') #A[1]

  def extract_max(self): # Implementa a opção 2 do menu do exercício, para remoção de elemento
    if self.m==0: #fila vazia
      return "-1 -1,0"
    else:
      max = self.A[1]
      self.A[1] = self.A[self.m]
      self.m = self.m-1
    self.max_heapify(self.A, 1)
    return str(max.id)+" "+str(float(max.prior.replace(',','.'))).replace('.',',')

  def get_parent(self, A, index): #método auxiliar do increase_key
    if ((index > 1) and (index < len(A))):
      return index//2
    return -1

  def increase_key(self, A, index, key): #método auxiliar do insert - insere  na folha e corrigi as posições comparando com o pai
    A[index] = key
    while((index>1) and (float(A[self.get_parent(A, index)].prior.replace(',','.')) < float(A[index].prior.replace(',','.')))):
      A[index], A[self.get_parent(A, index)] = A[self.get_parent(A, index)], A[index]
      index = self.get_parent(A, index)

  def insert(self, key): # Método que implementa a opção 3 do menu para inserção de elementos
    self.m = self.m+1
    self.increase_key(self.A, self.m, key)
    print("T")

  def HeapPrint(self): #Exibe a fila de prioridade para o comando 4 do exercício
    texto="";
    indice=0;
    for elemt in range(1,self.m+1):
        indice=+1
        texto=texto + str(self.A[elemt].id)+" "+str(float(self.A[elemt].prior.replace(',','.'))).replace('.',',') + " " if indice<self.m else ""
    print(texto)
```
